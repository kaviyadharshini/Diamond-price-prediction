```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv("diamonds.csv")
dataset
```

```python
df_dataset = pd.DataFrame()
n_cut ={'Ideal':0,'Premium':1,'Good':2,'Very Good':3,'Fair':4}
n_color ={'D':0,'E':1,'F':2,'G':3,'H':4,'I':5,'J':6}
n_clarity ={'I1':0,'IF':1,'SI1':2,'SI2':3,'VS1':4,'VS2':5,'VVS1':6,'VVS2':7}
```

```python
def clean_cut(m):
    return n_cut[m]
def clean_color(m):
    return n_color[m]
def clean_clarity(m):
    return n_clarity[m]
df_dataset['carat']=dataset['carat']
df_dataset['cut']=dataset['cut'].apply(clean_cut)
df_dataset['color']=dataset['color'].apply(clean_color)
df_dataset['clarity']=dataset['clarity'].apply(clean_clarity)
df_dataset['depth']=dataset['depth']
df_dataset['table']=dataset['table']
df_dataset['x']=dataset['x']
df_dataset['y']=dataset['y']
df_dataset['z']=dataset['z']
df_dataset['price']=dataset['price']
print(df_dataset.head(10))
```

```python
feature_names =['carat','cut','color','clarity','depth','table','x','y','z']
m =df_dataset[feature_names]
n =df_dataset.price
print(m)
print(n)
```

```python
from sklearn.model_selection import train_test_split
m_train,m_test,n_train,n_test =train_test_split(m,n,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(m_train,n_train)
pred =model.predict(m_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
test_set_r2 =r2_score(n_test,pred)
```

```python
print('R2 score:',test_set_r2)
plt.scatter(n_test,pred,color='red')
plt.title('Linear Regression')
plt.xlabel('Actual charges')
plt.ylabel('Predicted charges')
plt.show()
```
