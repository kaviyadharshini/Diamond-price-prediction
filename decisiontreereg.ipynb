```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('diamonds.csv')
dataset
```

```python
df_dataset = pd.DataFrame()
n_cut ={'Ideal':0,'Premium':1,'Good':2,'Very Good':3,'Fair':4}
n_color ={'D':0,'E':1,'F':2,'G':3,'H':4,'I':5,'J':6}
n_clarity ={'I1':0,'IF':1,'SI1':2,'SI2':3,'VS1':4,'VS2':5,'VVS1':6,'VVS2':7}
```

```python
def clean_cut(m):
    return n_cut[m]
def clean_color(m):
    return n_color[m]
def clean_clarity(m):
    return n_clarity[m]
df_dataset['carat']=dataset['carat']
df_dataset['cut']=dataset['cut'].apply(clean_cut)
df_dataset['color']=dataset['color'].apply(clean_color)
df_dataset['clarity']=dataset['clarity'].apply(clean_clarity)
df_dataset['depth']=dataset['depth']
df_dataset['table']=dataset['table']
df_dataset['x']=dataset['x']
df_dataset['y']=dataset['y']
df_dataset['z']=dataset['z']
print(df_dataset.head(10))
```

```python
m = df_dataset.iloc[:, 0:10].values
m
```

```python
n = dataset.iloc[:, 10].values
n
```

```python
from sklearn.model_selection import train_test_split
m_train, m_test, n_train, n_test = train_test_split(m,n,test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(m_train,n_train)
```

```python
test_score =regressor.score(m_test,n_test)
training_score =regressor.score(m_train,n_train)
train_error =1-training_score
print("training score:",training_score,"\ntraining error:",train_error)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
n_pred=regressor.predict(m_test)
r2_test=r2_score(n_test,n_pred)
```

```python
plt.scatter(n_test,n_pred,color='red')
plt.title('Decision Tree Regression')
plt.xlabel('actual charges')
plt.ylabel('predicted charges')
plt.show()
```

```python
print("R2 score for test dataset: ",r2_test)
print("test error:",1-r2_test)
```
